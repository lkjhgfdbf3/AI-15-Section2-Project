```python

df1 = pd.read_excel('/content/LOCALDATA_ALL_무인민원발급기발급제증명정보.xlsx',sheet_name='무인민원발급기발급제증명정보_1')
df2 = pd.read_excel('/content/LOCALDATA_ALL_무인민원발급기발급제증명정보.xlsx',sheet_name='무인민원발급기발급제증명정보_2')
df3 = pd.read_excel('/content/LOCALDATA_ALL_무인민원발급기설치정보.xlsx')
```

```python
df1_clean = df1.copy()
df2_clean = df2.copy()
df3_clean = df3.copy()
```

```python
result1 = pd.concat([df1_clean,df2_clean])
result1 = result1.reset_index(drop=True)
result1
```

```python
list_total_loc_name = df3_clean['관리기관명'].unique()
for i in list_total_loc_name:
  test1 = result1.loc[result1['시군구명']==i]
  arr1 = test1['발급기번호'].unique()
  test2 = df3_clean.loc[df3_clean['관리기관명']==i]
  arr2 = test2['관리번호'].unique()
  list_compare = [x for x in arr1 if x not in arr2]
  for i in list_compare:
    idx = test1.loc[test1['발급기번호']==i].index
    result1.drop(idx,inplace=True)
```

```python
result1 = result1.reset_index(drop=True)
result1
```

```python
result1['데이터갱신일자'] = result1['데이터갱신일자'].str.split().str[0]
```

```python
result1['갱신 연도']=result1['데이터갱신일자'].str.split('-').str[0]
result1['갱신 월']=result1['데이터갱신일자'].str.split('-').str[1]
result1['갱신 일']=result1['데이터갱신일자'].str.split('-').str[2]
```

```python
result1 = result1.drop(['번호','개방서비스ID','개방자치단체코드','관리번호','초기메뉴번호','민원사무분류번호','데이터갱신일자','최종수정일자'],axis=1)
```
